```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.animation import FuncAnimation

np.random.seed(42)
data_A = np.random.normal(loc=5, scale=1.0, size=1000)
data_B = np.random.normal(loc=6.5, scale=1.5, size=1000)

fig, ax = plt.subplots()

bins = np.linspace(0, 12, 30)

def update_hist(num, data_A, data_B, ax, bins):
    ax.clear()
    sns.histplot(data_A[:int(num)], kde=False, color="blue", bins=bins, label="Group A", ax=ax, stat="density")
    sns.histplot(data_B[:int(num)], kde=False, color="green", bins=bins, label="Group B", ax=ax, stat="density")
    ax.set_title('Distribution of Model Predictions Across Two Demographic Groups')
    ax.set_xlabel('Prediction Value')
    ax.set_ylabel('Density')
    ax.legend()

ani = FuncAnimation(fig, update_hist, frames=np.linspace(10, 1000, 100), fargs=(data_A, data_B, ax, bins))

ani.save('animated_bias_seaborn.gif', writer='pillow', fps=5)

plt.show()

anim_reference = ani
```
