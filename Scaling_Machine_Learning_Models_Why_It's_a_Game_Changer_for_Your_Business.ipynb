```python
import pandas as pd
import numpy as np
import plotly.express as px

# Simulating a dataset
np.random.seed(42)
dates = pd.date_range(start='2020-01-01', periods=24, freq='M')
data = {
    'Date': dates,
    'North_Region_Sales': np.random.randint(100, 500, size=(24)),
    'South_Region_Sales': np.random.randint(50, 400, size=(24)),
    'East_Region_Sales': np.random.randint(80, 450, size=(24)),
    'West_Region_Sales': np.random.randint(70, 500, size=(24))
}
df = pd.DataFrame(data)

# Melt the data to make it suitable for plotting
df_melted = df.melt(id_vars='Date', var_name='Region', value_name='Sales')

# Creating an animated bar chart
fig = px.bar(df_melted, x='Region', y='Sales', color='Region', animation_frame='Date',
             animation_group='Region', range_y=[0, 600], title='Dynamic Sales Performance by Region')

fig.show()
```
